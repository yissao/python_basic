```python
from youtube_transcript_api import YouTubeTranscriptApi

video_id = "PldyIefUPOg"

try:
    transcript_list = YouTubeTranscriptApi.list_transcripts(video_id)

    # Busca a transcrição gerada automaticamente em português
    transcript = transcript_list.find_generated_transcript(['pt'])

    # O resultado é um FetchedTranscript com .snippets
    data = transcript.fetch()

    # Itera corretamente sobre os FetchedTranscriptSnippet
    for entry in data.snippets:
        print(f"{entry.start:.2f}s: {entry.text}")

except Exception as e:
    print(f"Erro: {e}")
```

```python
full_text = " ".join([entry.text for entry in data.snippets])
print(full_text)
```

```python
# Configura a API Key do Google Gemini

import os
from google.colab import userdata

os.environ["GOOGLE_API_KEY"] = userdata.get('GOOGLE_API_KEY')
```

```python
# Configura o cliente da SDK do Gemini

from google import genai

client = genai.Client()

MODEL_ID = "gemini-2.0-flash"
```

```python
# Pergunta ao Gemini uma informação mais recente que seu conhecimento

from IPython.display import HTML, Markdown

# Perguntar pro modelo quando é a próxima imersão de IA ###############################################
resposta = client.models.generate_content(
    model=MODEL_ID,
    contents='corrija erros óbvios devido a geração automática de legenda. Mantenha o máximo do conteúdo do transcrição' + full_text,
)

# Exibe a resposta na tela
display(Markdown(f"Resposta:\n {resposta.text}"))
```

```python

#=========================================================
from google import genai

cliente = genai.Client()
#=========================================================
modelo = 'gemini-2.0-flash'
#=========================================================
# resposta = cliente.models.generate_content(model=modelo, contents="quantas pessoas foram a Lua?")
# # print(resposta.result) #erro AttributeError: 'GenerateContentResponse' object has no attribute 'result'
# print(resposta)
# #=========================================================
# print(resposta.text)

from google.genai import types

chat_config = types.GenerateContentConfig(
    system_instruction ="""
    Objetivo: Anotações para estudo
    Instruções: A seguir está a transcrição de um vídeo do YouTube.
    ⚠️ Atenção: a transcrição foi gerada automaticamente e pode conter erros de pontuação, palavras trocadas ou trechos desconexos.
    Sua tarefa é:
    Corrigir os trechos mal transcritos com base no contexto geral do vídeo.
    Organizar a informação em parágrafos coesos e claros.
    Produzir um resumo detalhado, destacando os principais tópicos abordados, exemplos dados, argumentos utilizados e qualquer conclusão relevante.
    Separe os termos técnicos e faça uma explicação breve deles
    Se possível, identifique o público-alvo ou a intenção do vídeo.
    """,
    temperature=0.5
)
chat=cliente.chats.create(model=modelo, config=chat_config)
resposta = chat.send_message(full_text)
display(Markdown(f"Resposta:\n {resposta.text}"))
```
