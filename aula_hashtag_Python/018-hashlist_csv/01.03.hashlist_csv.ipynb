```python
# pandas
import pandas as pd

verificar = 'hashlist10.txt' #nome do arquivo para colocar na variavel teste
Hist_A = 'Historico_A.txt' #nome do arquivo para colocar na variavel var_A
Hist_B = 'Historico_B.txt' #nome do arquivo para colocar na variavel var_B
colspecs = [(0,32),(32,34),(34,-1)]
var_A = pd.read_fwf(Hist_A,colspecs = colspecs, names=['hashmd5sum','sp','arquivo'] )
var_B = pd.read_fwf(Hist_B,colspecs = colspecs, names=['hashmd5sum','sp','arquivo'] )
teste = pd.read_fwf(verificar,colspecs = colspecs, names=['hashmd5sum','sp','arquivo'] )

display(teste)
display(var_A)
display(var_B)

#historico=pd.concat([var_A,var_B])
#historico = var_A.append(var_B)
historico = var_A.append(var_B, ignore_index = True)
display(historico)
```

```python
comparado_tem = teste[teste['hashmd5sum'].isin(list(historico['hashmd5sum']))]
display(comparado_tem)

comparado_naotem = teste[~teste['hashmd5sum'].isin(list(historico['hashmd5sum']))]
display(comparado_naotem)

comparado_tem['arquivo'].to_csv('listax.txt', index=False, header=False)
display(historico)
comparado_naotem['hashmd5sum'].to_csv('historico_C.txt', index=False, header=False, mode='a')
historico = pd.read_fwf('Historico_C.txt',colspecs = colspecs, names=['hashmd5sum','sp','arquivo'] )
display(historico)
```
