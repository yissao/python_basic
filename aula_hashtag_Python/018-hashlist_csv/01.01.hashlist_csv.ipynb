```python
# pandas
import pandas as pd

verificar = 'hashlist10.txt' #nome do arquivo para colocar na variavel teste
historicoarquivo = 'Historico.txt' #nome do arquivo para colocar na variavel historico
colspecs = [(0,32),(32,34),(34,-1)]
historico = pd.read_fwf(historicoarquivo,colspecs = colspecs, names=['hashmd5sum','sp','arquivo'] )
teste = pd.read_fwf(verificar,colspecs = colspecs, names=['hashmd5sum','sp','arquivo'] )

display(teste)

comparado_tem = teste[teste['hashmd5sum'].isin(list(historico['hashmd5sum']))]
display(comparado_tem)

comparado_naotem = teste[~teste['hashmd5sum'].isin(list(historico['hashmd5sum']))]
display(comparado_naotem)
```

```python
import os
import hashlib
from tqdm import tqdm #pip3 install --target="C:\Program Files\Python310\Scripts" tqdm ;cmd como ADM

path = '/Users/raziel/Python/018-hashlist_csv'
files = os.listdir(path)
# pbar = tqdm(total=len(path), position=0, leave=True)

for cadaarquivo in files:
    md5_hash = hashlib.md5()
    with open(cadaarquivo,"rb") as f:
        # Read and update hash in chunks of 4K
        for byte_block in iter(lambda: f.read(4096),b""):
            md5_hash.update(byte_block)
        print(md5_hash.hexdigest())
```
